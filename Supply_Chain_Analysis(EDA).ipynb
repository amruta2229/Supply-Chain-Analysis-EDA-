```python
# Load data from a CSV file
data = pd.read_csv('supply_chain_data.csv')

fig = px.scatter(
    data, 
    x='Price', 
    y='Revenue generated', 
    color='Product type', 
    hover_data=['Number of products sold'], 
    trendline="ols"
)
fig.show()
```

```python
sales_data = data.groupby('Product type')['Number of products sold'].sum().reset_index()

sales_data
```

```python
pie_chart=px.pie(sales_data , values='Number of products sold', names = 'Product type',title = 'Sales by Product Type',hover_data=['Number of products sold'],hole =0.5)
pie_chart.update_traces(textposition='inside', textinfo='percent+label')
pie_chart.show()
```

```python
Customer_data = data.groupby('Customer demographics')['Number of products sold'].sum().reset_index()

pie_chart = px.pie(Customer_data,values='Number of products sold', names='Customer demographics',
                  title='Sales by Customer',
                  hover_data=['Number of products sold'],
                  hole=0.5,
                  color_discrete_sequence=px.colors.qualitative.Pastel
                  )
pie_chart.update_traces(textposition='inside',textinfo='percent+label')
pie_chart.show()
```

```python
total_revenue = data.groupby('Shipping carriers')['Revenue generated'].sum().reset_index()
total_revenue
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure()

fig.add_trace(go.Bar(x=total_revenue['Shipping carriers'],y=total_revenue['Revenue generated']))
fig.update_layout(title='Total Revenue by Shipping Carrier'
                 ,xaxis_title= 'Shipping Carrier' 
                 ,yaxis_title= 'Revenue Generated')

fig.show()
```

```python
total_revenue = data.groupby('Location')['Revenue generated'].sum().reset_index()
fig = go.Figure()
fig.add_trace(go.Bar(x = total_revenue['Location'] ,y = total_revenue['Revenue generated']))
fig.update_layout(title = 'Total Revenue based on location' 
                 ,xaxis_title= 'Locations'
                 , yaxis_title= 'Revenue generated')
fig.show()
```

```python
avg_lead_time = data.groupby('Product type')['Lead time'].mean().reset_index()
avg_manufacturing_costs = data.groupby('Product type')['Manufacturing costs'].mean().reset_index()
result = pd.merge(avg_lead_time,avg_manufacturing_costs , on = 'Product type')
result.rename(columns={'Lead time': 'Average Lead Time', 'Manufacturing costs': 'Average Manufacturing Costs'}, inplace=True)
result
```

```python
revenue_chart = px.line(data ,x = 'SKU',y='Revenue generated',title ='Revenue Generated by SKU')
revenue_chart.show()
```

```python
order_quantity_chart = px.bar(data,x='SKU', 
                              y='Order quantities', 
                              title='Order Quantity by SKU')


order_quantity_chart.show()
```

```python
shipping_cost_chart = px.bar(data, x='Shipping carriers', 
                             y='Shipping costs', 
                             title='Shipping Costs by Carrier')
shipping_cost_chart.show()
```

```python
transportation_chart =px.pie(data,names = 'Transportation modes' ,values = 'Costs' 
                             ,title ='Cost Distribution by Transportation Mode',
                             hole =0.5 ,
                             color_discrete_sequence = px.colors.qualitative.Pastel)
transportation_chart.show()
```

```python
defect_rates_by_product = data.groupby('Product type')['Defect rates'].mean().reset_index()
defect_rates_by_product
```

```python
figure = px.bar(defect_rates_by_product,x= 'Product type' ,y='Defect rates',title ='Average Defect Rates by Product Type') 

figure.show()
```

```python
pivot_table  = pd.pivot_table(data,values='Defect rates',index=['Transportation modes'],aggfunc = 'mean')
pivot_table
```

```python
transportation_chart =px.pie(values=pivot_table["Defect rates"], 
                              names=pivot_table.index,  
                             title= 'Defect Rates by Transportation Mode' , 
                             color_discrete_sequence =  px.colors.qualitative.Pastel,
                             hole =0.5)

transportation_chart.show()
```
